```python
# Import Pandas for data manipulation
import pandas as pd
```

```python
# Open the text file and read all lines
with open('Sentences_75Agree_sample.txt', 'r', encoding='latin1') as file:
    lines = file.readlines()
lines[:5]
```

```python
# Split each line at '@' to separate sentences and labels and strip whitespace
data = [line.strip().split('@') for line in lines]
data[:5]
```

```python
# Define column names and create a DataFrame
df = pd.DataFrame(data, columns=['Sentence', 'Label'])
```

```python
# Display the top rows of the DataFrame to ensure it is structured correctly
df.head()
```

```python
# Import matplotlib for plotting
import matplotlib.pyplot as plt
```

```python
# Creatlabels_dfe a DataFrame specifically for the 'Label' column for easy manipulation
labels_df = pd.DataFrame(df.Label, columns=['Label'])
labels_df[:5]
```

```python
# Use a bar chart to visualize how frequently each label occurs in the dataset
labels_df['Label'].value_counts().plot(kind='bar')
plt.title('Distribution of Labels')  # Add a title to the bar chart
plt.xlabel('Label')                  # Label the x-axis as 'Label'
plt.ylabel('Count')                  # Label the y-axis as 'Count'
plt.show()                           # Display the plot
```

```python
# Determine the number of words in each sentence to analyze sentence length
sentence_lengths = [len(sentence.split()) for sentence in df.Sentence]
sentence_lengths[:5]
```

```python
# Plot a histogram to show the distribution of sentence lengths in the dataset
plt.hist(sentence_lengths, bins=30)
plt.title('Distribution of Sentence Lengths')  # Add a title to the histogram
plt.xlabel('Sentence Length')                  # Label the x-axis as 'Sentence Length'
plt.ylabel('Frequency')                        # Label the y-axis as 'Frequency'
plt.show()                                     # Display the plot
```
